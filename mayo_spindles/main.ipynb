```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Load annotations into a pandas dataframe

```python
df = pd.read_csv('sub-MH1_ses-EMU1_merged\M1_sleep_sleepspindles_011123.csv')
```

```python
pd.unique(df['Annotation'])
```

Plot duration distributions

```python
plt.figure(figsize=(10, 6))
sns.stripplot(x='Annotation', y='Duration', data=df, color='black', alpha=0.5)
plt.title('Distributions of Duration for Each Class')
plt.xticks(rotation=45)
plt.show()
```

Plot the difference between end-start and duration (expected)

```python
df['error'] = abs(df['Duration'] - (df['End'] - df['Start']))
sns.violinplot(x='error', data=df)
plt.title('Duration error distribution')
plt.show()
```
